```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
l = np.asarray([3,5])

# computing the gaussian kernel
def f(x,y,sigma_sq):
    X = [x,y]
    return np.exp( (-0.5/sigma_sq)*(np.linalg.norm(X-l)**2) )

# create 2D repr for x,y both ranging from 1 to 10
x = np.arange(-5,10,0.1)
y = np.arange(0,15,0.1)
X,Y = np.meshgrid(x,y)

# np.ravel(X) = 1D rearrangement
zs  = np.array([f(x,y,sigma_sq=1) for x,y in zip(np.ravel(X), np.ravel(Y))])
zs2 = np.array([f(x,y,sigma_sq=3) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z   = zs.reshape(X.shape)
Z2  = zs2.reshape(X.shape)
#print(Z[0])

# ploting
fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X,Y,Z)
ax1.set_title("Sigma Square = 1")
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X,Y,Z2)
ax2.set_title("Sigma Square = 3")
plt.show()
```
