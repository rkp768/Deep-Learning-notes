```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
l = np.asarray([3,5])

# computing the gaussian kernel
def f(x,y):
    X = [x,y]
    # sigma_sq = 1
    return np.exp( -0.5*(np.linalg.norm(X-l)**2))

# create 2D repr for x,y both ranging from 1 to 10
x = np.arange(-5,10,0.1)
y = np.arange(0,15,0.1)
X,Y = np.meshgrid(x,y)

# np.ravel(X) = 1D rearrangement
zs = np.array([f(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)
#print(Z[0])

# ploting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
```
