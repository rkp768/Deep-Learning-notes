```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate random sample data
N = 500 # number of samples per class
k = 3 # number of classes
D = 2 # dimensionality
X = np.zeros((N*k,D)) # initialize
y = np.zeros(N*k) # --do--
r = np.linspace(0.0,30.0,N) # fixed radius

np.random.seed(0)

for i in xrange(k):
    ix = range(i*N,N*(i+1))
    t = np.linspace(i*40,(i+1)*40,N)+np.random.randn(N)*0.3 #theta
    X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
    y[ix] = i
fig = plt.figure()
plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.Spectral)
plt.xlim([-3,3])
plt.ylim([-3,3])
plt.show()
```

```python
# generate random sample data
N = 100 # number of samples per class
k = 3 # number of classes
D = 2 # dimensionality
X = np.zeros((N*k,D)) # initialize
y = np.zeros(N*k) # --do--

np.random.seed(0)

for i in xrange(k):
    ix = range(i*N,N*(i+1))
    t = np.linspace(i*50,(i+1)*50,N) #theta
    tt = np.random.uniform(i,i+1,N)*0.35 # for 1st class:random numbers will be 
                                         # between (0,1), for 2nd class (1,2) so on
    X[ix] = np.c_[tt*np.sin(t),tt*np.cos(t)]
    y[ix] = i
fig = plt.figure()
plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.Spectral)
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.show()
```

```python
# generate random sample data
N = 100 # number of samples per class
k = 2 # number of classes
D = 2 # dimensionality
X = np.zeros((N*k,D)) # initialize
y = np.zeros(N*k) # --do--

np.random.seed(0)

# multiplication factor
arr = [[1,1],[1,-1],[-1,-1],[-1,1]]
for i in xrange(k):
    ix = range(i*N,N*(i+1))
    tt = np.random.uniform(0,0.8,[N,2])
    # simply scatter the same class points in opposite quads
    for j in ix:
        if j%2==0:
            X[j] = tt[j-i*N]*arr[i]  
        else:
            X[j] = tt[j-i*N]*arr[i+2]
    y[ix] = i
fig = plt.figure()
plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.Spectral)
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.show()
```
